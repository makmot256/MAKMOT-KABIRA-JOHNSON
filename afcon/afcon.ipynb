```python
import pandas as pd

# Read the CSV file
df = pd.read_csv("AfricaCupofNationsMatches.csv")
```

```python
df.head(7)
```

```python
# Create two new rows as dictionaries
new_rows = [
    {
        'Year': 2025, 'Date ': '1-Jan-25', 'Time ': '18:00', 'HomeTeam': 'Kenya', 'AwayTeam': 'Uganda',
        'HomeTeamGoals': 2, 'AwayTeamGoals': 2, 'Stage': 'Group A', 'SpecialWinConditions': '',
        'Stadium': 'Nairobi Stadium', 'City': 'Nairobi', 'Attendance': 50000, 'Tournament': 'AFCON'
    },
    {
        'Year': 2025, 'Date ': '2-Jan-25', 'Time ': '21:00', 'HomeTeam': 'Nigeria', 'AwayTeam': 'Ghana',
        'HomeTeamGoals': 1, 'AwayTeamGoals': 3, 'Stage': 'Group B', 'SpecialWinConditions': '',
        'Stadium': 'Lagos Stadium', 'City': 'Lagos', 'Attendance': 60000, 'Tournament': 'AFCON'
    }
]

# Append to DataFrame
df = pd.concat([df, pd.DataFrame(new_rows)], ignore_index=True)
df.tail(3)
```

```python
df['Tournament'] = 'AFCON'
df.head(3)
```
